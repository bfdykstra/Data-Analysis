This is just a simple assignment where we estimated a linear regression and then plotted it using the iris dataset.

```python
from sklearn import datasets
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 18.5, 10.5
import numpy as np

iris = datasets.load_iris()
from __future__ import division
```

```python
x = iris.data[:,1]
y = iris.data[:, 3]
plt.scatter(x,y);
```

After loading the data and getting an idea of what it looks like, we estimate a simple linear equation

```python
y_hat = -.75*x + 3.5
plt.plot(x, y_hat);
plt.show();
```

Now, we calculate the mean squared error between our predictions and the actual data

```python
mse_y = 0
for i in range(len(x)):
    #mse_x += (y_hat[i] - x[i])**2
    mse_y += (y_hat[i] - y[i])**2

mse = (mse_y)/len(x)
print(mse)
```

We now try to classify the different types of iris' based on a very simple equation

```python
#target 2 = virginica 
#target 0 = setosa
petal_length = iris.data[:,2]
petal_width = iris.data[:,3]
#plt.scatter(petal_width, petal_length)
plt.scatter(petal_width[iris.target == 2],petal_length[iris.target == 2], color = 'red');
plt.scatter(petal_width[iris.target == 0],petal_length[iris.target == 0], color = 'blue');
plt.plot(petal_width, -petal_width + 4, color = 'black');
```

Try a slightly different equation with a different intercept. This will misclassify a couple of samples

Get the classification error, ends up being 10 %

```python
y_hat = -petal_width + 2
y = petal_length[iris.target == 0]
count = 0
for i in range(len(petal_length[iris.target == 0])):
    if y_hat[i] < y[i]:
        count += 1

ce = count / len(y)
print ce
```
